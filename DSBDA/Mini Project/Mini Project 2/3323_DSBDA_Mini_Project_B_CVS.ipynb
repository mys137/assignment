```python
# Mini Project 2
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
url="covid_vaccine_statewise.csv"
df_vaccine=pd.read_csv(url)
```

```python
df_vaccine.shape
```

```python
df_vaccine.info()
```

```python
df_vaccine.describe()
```

```python
df_vaccine.drop('Total Individuals Vaccinated',axis=1,inplace = True)

#Making new column of Total Vaccinatons in dataset
df_vaccine['Total Vaccinatons'] = df_vaccine['First Dose Administered']+df_vaccine['Second Dose Administered']

#Renaming columns
df_vaccine.rename(columns = {'Updated On':'Date'}, inplace = True)
```

```python
df_vaccine.head()
```

```python
plt.figure(figsize=(15,30))
plt.bar( df_vaccine['State'],df_vaccine['First Dose Administered'] ,color ='C0')
plt.xticks(rotation='vertical')
plt.xlabel("States")
plt.ylabel("Count")
plt.title("Count of first dose administered state-wise")
plt.show()

```

```python
plt.figure(figsize=(15,30))
plt.bar( df_vaccine['State'],df_vaccine['Second Dose Administered'] ,color ='C1')
plt.xticks(rotation='vertical')
plt.xlabel("States")
plt.ylabel("Count")
plt.title("Count of second dose administered state-wise")
plt.show()

```
